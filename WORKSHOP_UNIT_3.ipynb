## i)Apply Feature Encoding for ordinal data 

```python
import pandas as pd
import numpy as np
import seaborn as sbn
df = pd.read_csv("/content/Workshop_feature Engg.csv")
df1 = df.copy()
df1.head(5)
```

```python

from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
aspiration = ['std','turbo']
enc = OrdinalEncoder(categories = [aspiration])
enc.fit_transform(df1[["aspiration"]])
```

```python
df1['aspiration'] = enc.fit_transform(df1[["aspiration"]])
df1.head(5)
```

## ii)Apply Feature Encoding for nomial data 

```python
df2 = df1.copy()
le = LabelEncoder()
df2['drivewheel'] = le.fit_transform(df2[["drivewheel"]])
df2.head(5)
```

## iii)Apply Feature Scaling for numerical dat

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
df1 = pd.read_csv("Workshop_feature Engg.csv")
X = df1[['carheight','curbweight']]
df1=sc.fit_transform(X)
print(df1)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
df1 = pd.read_csv("Workshop_feature Engg.csv")
X = df1[['carheight','curbweight']]
df1=mms.fit_transform(X)
print(df1)
```

```python
from sklearn.preprocessing import MaxAbsScaler
mas = MaxAbsScaler()
df1 = pd.read_csv("Workshop_feature Engg.csv")
X = df1[['carheight','curbweight']]
df1=mas.fit_transform(X)
print(df1)
```

```python
from sklearn.preprocessing import RobustScaler
rs = RobustScaler()
df1 = pd.read_csv("Workshop_feature Engg.csv")
X = df1[['carheight','curbweight']]
df1=rs.fit_transform(X)
print(df1)
```
